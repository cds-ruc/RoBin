```python
# There are some edge point in fb dataset
# edge point is
# need to confirm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 7, figsize=(35, 12))  # 创建一个3行7列的子图布局
axs = axs.flatten()  # 将子图的数组扁平化，以便于在循环中使用

pos=0
for dataset in ["fb"]:
    for test_suite in [99,2,4]:
        for bulk_size in [1000000,2000000,5000000,10000000,20000000,50000000,100000000]:
            file_name="/root/workspace/GRE/datasets/expr/"+dataset+"_"+str(test_suite)+"_"+str(bulk_size)+"_bulk.dump"
            data=np.fromfile(file_name, dtype=np.uint64)
            axs[pos].set_title(f'dataset={dataset}, test_suite={test_suite}, bulk_size={bulk_size}')  # 为子图添加标题
            if data[-1]>1e11:
                print(f'dataset={dataset}, test_suite={test_suite}, bulk_size={bulk_size}, max={data[-1]}')
            sns.ecdfplot(data[data<1e11], ax=axs[pos])
            pos+=1
```
