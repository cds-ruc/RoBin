```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Example data
np.random.seed(0)
data = np.fromfile("/root/workspace/GRE/datasets/fb", dtype=np.uint64)

# Sort the data
sorted_data = np.sort(data)
ecdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Create figure and axis
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, gridspec_kw={'width_ratios': [10, 1], 'wspace': 0.05})
fig.set_size_inches(12, 5)
# Plot the main data without the long tail
main_data = sorted_data[sorted_data < 140805133574147]
main_ecdf = ecdf[sorted_data < 140805133574147]
ax1.step(main_data, main_ecdf, where='post')

# Plot the long tail data
long_tail_data = sorted_data[sorted_data >= 140805133574147]
long_tail_ecdf = ecdf[sorted_data >= 140805133574147]
ax2.step(long_tail_data, long_tail_ecdf, where='post')

# Add labels and titles
ax1.set_xlabel('Data')
ax2.set_xlabel('Long Tail')
ax1.set_ylabel('ECDF')
fig.suptitle('ECDF with Broken X-Axis')

# Customize the plots to indicate a break
ax1.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax1.yaxis.tick_left()
ax2.yaxis.tick_right()
ax2.tick_params(left=False)

# Add diagonal lines to indicate break
d = .015  # How big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((1-d, 1+d), (-d, +d), **kwargs)
ax1.plot((1-d, 1+d), (1-d, 1+d), **kwargs)


# 1e11. 
kwargs.update(transform=ax2.transAxes)  # Switch to the right axes
ax2.plot((-d, +d), (1-d, 1+d), **kwargs)
ax2.plot((-d, +d), (-d, +d), **kwargs)

plt.show()
```
